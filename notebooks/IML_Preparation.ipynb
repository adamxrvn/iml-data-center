```python
def transform_dataset(df1):
  df1['Timestamp'] = pd.to_datetime(df1['Timestamp'])
  df1.sort_values(['Timestamp'], ascending=True, inplace=True)
  df1.set_index('Timestamp', inplace=True)
  return df1
```
